```python
import pandas as pd
from google.colab import drive
drive.mount('/drive')
pd.set_option('display.max_columns',12)
pessoas = pd.read_csv("/drive/MyDrive/class_titanic.csv")
display(pessoas[0:13])
```

```python
colunaIdade= pessoas["Age"]
nova_colunaIdade= colunaIdade.dropna()
display(nova_colunaIdade[0:13])
```

```python
media=nova_colunaIdade.mean()
print(media)
```

```python
total=pessoas.loc[(pessoas['Age'])< media+5.5]
resultado=total.loc[(pessoas['Age'])> media-5.5]
resultado_final=resultado.dropna()
display(resultado_final[0:13])

```

```python
resultado_5colunas=resultado_final[['PassengerId','Pclass','Name','Sex','Age']]
display(resultado_5colunas[0:13])
```

```python
import random
coluna1=(pessoas['Age']-pessoas['Pclass'])*random.random()
print(coluna1)
```

```python
import numpy as np
pessoas1 = pessoas.reindex(np.random.permutation(pessoas.index))

pd.set_option('display.max_columns', 10)
pd.set_option('display.max_rows', 15)
display(pessoas1[0:13])
```

```python
#Ordenando os itens pela ordem alfabética
pessoas2 = pessoas.sort_values(by='Name', ascending=True)
print(f"A primeira pessoa é: {pessoas2['Name'].iloc[0]}")
      
pd.set_option('display.max_columns', 12)
pd.set_option('display.max_rows', 5)

display(pessoas2)

```

```python
sobreviventes = pessoas.groupby('Survived')[['Name']].count()
display(sobreviventes)
```

```python
#Critério: classicar o Pclass(classe do passageiro) em classe econômica, classe econômica premium e primeira classe seguindo os critérios do dicionario abaixo.
pessoas['Pclass'] = pessoas['Pclass'].map(
    {3:"Classe econômica", 2: "Classe econômica premium", 1: "Primeira classe"})
pd.set_option('display.max_columns', 7)
pd.set_option('display.max_rows', 10)
display(pessoas)
```
